```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
url = '/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/insurance_mod.csv'
df = pd.read_csv(url)
df.head()
```

```python
ax = df['bmi'].hist(bins = 20, edgecolor = 'black')
ax.set(ylabel='Number of Individuals', xlabel='BMI', title='Distribution of BMI')
```

```python
ax2 = sns.boxplot( x ='region',y ='bmi', data=df)
ax2.set(title='BMI By Region')
```
